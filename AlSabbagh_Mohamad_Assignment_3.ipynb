```python
# Task 1 - Images with Python (OpenCV)

import cv2
import matplotlib.pyplot as plt

# 1. Load a different image and display it
image = cv2.imread("newborn.jpg")  

if image is None:
    print("Error: Image not found.")
else:
    # Convert to RGB for correct display
    image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

    plt.imshow(image_rgb)
    plt.title("Original Image")
    plt.axis("off")
    plt.show()

    # 2. Convert image to grayscale and display it
    gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    
    plt.imshow(gray_image, cmap="gray")
    plt.title("Grayscale Image")
    plt.axis("off")
    plt.show()
```

```python

from PyPDF2 import PdfReader, PdfWriter

# 1. Merge two PDF files 
reader1 = PdfReader("Turtle.pdf")
reader2 = PdfReader("Rabbit.pdf")

writer = PdfWriter()

# Add pages from first PDF
for page in reader1.pages:
    writer.add_page(page)

# Add pages from second PDF
for page in reader2.pages:
    writer.add_page(page)

# Save merged PDF
with open("Merged_Animals.pdf", "wb") as output_file:
    writer.write(output_file)

print("Merged PDF created successfully.")


# 2. Extract text from all pages 
reader = PdfReader("Sample.pdf")

print("\nNumber of pages:", len(reader.pages))
print("\nExtracted Text:\n")

for i, page in enumerate(reader.pages, start=1):
    text = page.extract_text()
    print(f"--- Page {i} ---")
    
    if text:
        print(text)
    else:
        print("No text found on this page.")
    
    print()
```

```python
from flask import Flask, request, jsonify
from twilio.rest import Client

# Twilio Setup

account_sid = "XXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXX" # for security reasons
auth_token = "XXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXX" # for security reasons
client = Client(account_sid, auth_token)

# --------------------------
# Recipients inside the code

recipients = [
    "+16132403573",
    "+16132403573"    
]

# Function to send bulk SMS
def send_bulk_sms(message):
    for number in recipients:
        sms = client.messages.create(
            body=message,
            from_="+15799772464",   #Twilio number
            to=number
        )
        print(f"Message sent to {number} with SID: {sms.sid}")

# Flask App
app = Flask(__name__)

@app.route("/")
def home():
    return "SMS Flask App Running!"

@app.route("/send_sms")
def send_sms():
    message = "Hello, this is a test SMS sent to multiple recipients!"
    send_bulk_sms(message)
    return jsonify({"status": "SMS sent to all recipients"})


# Run Flask
if __name__ == "__main__":
    app.run(debug=True, use_reloader=False)
```
